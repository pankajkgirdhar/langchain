```python
from pydantic import BaseModel, Field
from langchain.agents import create_agent
from langchain_groq  import ChatGroq
import os
from dotenv import load_dotenv
load_dotenv()

```

```python
class ContactInfo(BaseModel):
    name: str
    email: str
    phone: str

class ContactInfoList(BaseModel):
    contacts: list[ContactInfo] = Field(..., description="List of contacts")
```

```python
model = ChatGroq(model="openai/gpt-oss-20b")
agent = create_agent(model
                    , tools=[]
                    , response_format=ContactInfoList)


result = agent.invoke({"messages":" extract info from: pank.89@gmail.com, 9588716564, Pankaj, Amaan Girdhar, amaa@yahoo.com, 6586586545"})

print(result["messages"][-1].content)

```
