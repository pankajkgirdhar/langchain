```python
from langchain.agents import create_agent, AgentState
from langchain.messages import SystemMessage
from langchain_ollama import ChatOllama
from langchain_groq  import ChatGroq
import custom_tools
```

```python
basic_model = ChatOllama(model="mistral:7b",base_url="http://localhost:11434",temperature=0.0)
advanced_model = ChatGroq(model="openai/gpt-oss-20b", temperature=0.0)

sys_msg = SystemMessage(content="You are a zero helpful assistant that can use tools to answer questions.")

tool_list = [custom_tools.search_internet, custom_tools.calculate, custom_tools.summarize_conversation]
agent = create_agent(
    model=advanced_model,  # Default model
    tools=tool_list,
    system_prompt=sys_msg,
)

result = agent.invoke({"messages": "What is capital of India?"})

print(result["messages"][-1].content)
```

```python
basic_model = ChatOllama(model="mistral:7b",base_url="http://localhost:11434",temperature=0.0)
advanced_model = ChatGroq(model="openai/gpt-oss-20b", temperature=0.0)

tool_list = [custom_tools.search_internet, custom_tools.calculate, custom_tools.summarize_conversation]
agent = create_agent(
    model=advanced_model,  # Default model
    tools=tool_list,
    system_prompt="You are a wasteful helpful assistant that can use tools to answer questions.",
)

result = agent.invoke({"messages": [{"role":"system","content":"You are a total waste assistant that always give wrong answer."},{"role":"user","content":"What is capital of India?"}]})

print(result)
print(result["messages"][-1].content)
```
